```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)

def Lennard_Jones(r, eps, sigma):
    sig_r6 = (sigma/r)**6
    return 4*eps*(sig_r6**2 - sig_r6)

def Lennard_Jones_curv(r, eps, sigma):
    return 4*eps*(42*sigma**6*(1/r)**8 - 156*sigma**12*(1/r)**14)

res = 100
noise = 0.001
damp = 10/r**3

r = np.linspace(0.95,2.,res)
LJ = Lennard_Jones(r, 1, 1)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,9))

# ax.plot(r, LJ, c="xkcd:orange", linewidth=3)
for i in range(15):
    ax.plot(r, LJ + np.cumsum(np.cumsum([(noise*rnd - noise/2) for rnd in np.random.rand(res)]))*damp, c="xkcd:orange", alpha = 0.35 + 0.01*i, linewidth=2)
    ax.plot(r, LJ + np.cumsum(np.cumsum([(noise*rnd - noise/2) for rnd in np.random.rand(res)]))*damp, c="xkcd:orange", alpha = 0.35 + 0.01*i, linewidth=2)
    
ax.set_facecolor('xkcd:navy blue')
ax.set_ylim(top=1.5)
ax.set_xlim(right=2)
ax.set
ax.set_xticks([])
ax.set_yticks([])

fig.tight_layout()
fig.savefig("/home/thism292/Documents/Data/CCS/logo_background.png", dpi=300)



LJ_curv = Lennard_Jones_curv(r, 1, 1)
fig2, ax2 = plt.subplots(nrows=1, ncols=1, figsize=(16,9))
ax2.plot(r, LJ_curv, c="xkcd:orange", linewidth=3)
ax2.set_ylim(-8, 8)

sw = np.argmin(np.abs(LJ_curv))
print(sw)
print(r[sw])
```
